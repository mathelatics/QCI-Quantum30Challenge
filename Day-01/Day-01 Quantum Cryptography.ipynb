> There are multiple types of attacks are possible either direct or indirect but here we are only concern with the indirect one becouse it's not our field to go into direct attacks.

1. **Trojan Horse Attack**
> It is a stand alone program which is usefull in the software unitility and games but it contains some hidden malicious code that can affect our computer systems 
- A  Trojan  horse  program,  when  executed,  may  change  the  invoking  users  file permissions so that the files are readable by any user.
- A Trojan horse may insert additional code into certain programs as they are compiled. For example, it may create a backdoor in the login program that permits the author of the Trojan horse to log on to the system using a special password.
- A Trojan horse may carry other malicious programs as its payload and install them when it is executed.
- A Trojan horse may also cause deletion, modification, copying of data, disrupting performance of computer system or replacing existing files with malicious files.

**Trapdoor Attack(Backdoor Attack)**
> It is a secret entry point into a program that allows someone who is aware of the trapdoor to gain access without going through the usual security access (authentication) procedures. Programmers use backdoors to debug and test programs. To debug the program, the developer may wish to gain special privileges and/or may wish to avoid long setup and authentication procedure. The programmer may also want to ensure that there is a method of activating the program should something be wrong with the authentication procedure itself. The backdoor is the code that recognizes some special sequence of input or is triggered by being run from a certain user ID or by an unlikely sequence of events etc.

**Stack and Buffer Overflow**
> When you started learning c programming there is a concept of overflow(overloading etc). A buffer overflow occurs when in a running program or process there is an attempt to write more data to a fixed length block of memory, or buffer, than the size of memory allocated to the buffer. Since buffers are created to contain a fixed amount of data, the extra data can overwrite  data  values  in  memory  addresses  adjacent  to  the  destination  buffer  unless  the program includes sufficient bounds checking to raise a flag or discard data when input data size exceeds the size of the memory buffer. Exploiting a buffer overflow allows an attacker to control or crash the process or to modify its internal variables
